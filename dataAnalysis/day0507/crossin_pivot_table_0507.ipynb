```python
# 1 将上方数据表建立 DataFrame 进行存储（阅读参考文章①）

import pandas as pd
import numpy as np
# datalist = pd.date_range(start='2019-1-01', end='2019-1-31')
df = pd.DataFrame({
    'id': [1001,1001,1002,1003,1001,1002,1002,1003,1001,1002,1001,1002],
    'date': ['2019-01-01', '2019-01-01', '2019-01-01','2019-01-01', '2019-01-02','2019-01-02', '2019-01-02', '2019-01-02', '2019-01-03', '2019-01-03', '2019-01-04', '2019-01-04'],
    'goods':['A','B','A','B','A','A','B','A','A','B','A','A'],
    'sales': [3000, 2000, 3000, 2000, 3000, 3000, 2000,3000, 3000, 2000,3000, 3000]},
    columns = ['id','date','goods','sales'])
df
```

```python
# 2 利用 pivot_table 统计该企业每个人每天销售商品 A 和 B 的金额之和（阅读参考文章②④⑤）

# pd.pivot_table(df, index="id", columns="date", values="sales", aggfunc=sum）

pd.pivot_table(df, index=['id','goods'], columns='date', values='sales', aggfunc=[np.sum, np.mean], fill_value=0)
```

```python
# 3 请将上面的统计数据再增加一列，统计出每个人的销售总金额（阅读参考文章②④⑤）

'''
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html

margins:  bool, default False
          Add all row / columns (e.g. for subtotal / grand totals).
margins_name: str, default ‘All’
              Name of the row / column that will contain the totals when margins is True.
'''
pd.pivot_table(df, 
               index=['id'], 
               columns=['date'], 
               values=['sales'], 
               fill_value=0, 
               aggfunc=sum, 
               margins=True,
               margins_name='每人销售总金额')
```

```python
# 讲数据存入 DataFrame 中
df1 = pd.DataFrame({'sex':['man','man','women','women','man','women','man','women','women'],
                    'age':[15,23,25,17,35,57,24,31,22],
                    'smoke':[True,False,False,True,True,False,False,True,False],
                    'height':[168,179,181,166,173,178,188,190,160]},
                    columns = ['sex','age','smoke','height'])
df1
```

```python
# 请以性别为分组依据，查看男女抽烟平均年龄和身高

pd.pivot_table(df1,  index=["sex"] ,  values=['smoke','age','height'], aggfunc=[np.mean])
```

```python
'''
aggfunc: function, list of functions, dict, default numpy.mean
If list of functions passed, the resulting pivot table will have hierarchical columns whose top level are the function names (inferred from the function objects themselves) If dict is passed, the key is column to aggregate and value is function or list of functions.

'''
#  因为aggfunc默认的方法就是 numpy.mean 求平均值，所以可以直接这样简单写

df1.pivot_table(index='sex')
```

```python
# 利用交叉表知识，统计各个性别抽烟的人数

'''
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html

pandas.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, margins_name: str = 'All', dropna: bool = True, normalize=False) 

index:   在结果透视表的行上进行分组的列名分组键
columns: 在结果透视表的列上进行分组的列名分组键
values:  需要聚合的列名，默认所有数值的列
rownames: 给index起名
colnames：给columns起名
aggfunc:  聚合函数或函数列表,默认计算频率表
margins:  添加行列总计(与aggfunc对应)
margins_name: 与margins对应
dropna:       是否除去均为nan的列

'''
pd.crosstab(index=df1.smoke, columns=df1.sex)
```

```python
# 或者 交换下位置
pd.crosstab(index=df1.sex, columns=df1.smoke)
```

```python
# 利用交叉表知识，统计各个年龄段抽烟人情况
pd.crosstab(index=df1.smoke, columns=df1.age)
```

```python
# 或者：交换下位置

pd.crosstab(index=df1.age, columns=df1.smoke)
```
