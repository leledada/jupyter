```python
'''
任务一：
请根据上方表格数据创建一个 DataFrame 存储公司的数据，数据命名为 df_company（阅读参考文章①）
https://zhuanlan.zhihu.com/p/37856914

'''

import pandas as pd
df_company = pd.DataFrame({'marketValue': [449, 371, 237, 21313, 1369, 823],
                                  'peratio': [8.31, 15.36, 16.01, 7.16, 7.59, 6.3],
                                'code': ['600926', '002958', '601128', '601398', '601229', '600919'],
                            'company': ['杭州银行', '青农商行', '常熟银行', '工商银行', '上海银行', '江苏银行']
                           },columns = ['company','code','marketValue', 'peratio'])
df_company
```

```python
df_company.set_index('code',drop=False)
```

```python
'''
任务二：
选出市值低于 2000 亿的所有公司（任务提示：使用布尔形式选取数据，阅读参考文章②③）
'''
# df_company
df_company[df_company.marketValue < 2000]
```

```python
'''
任务三：
选出市值 < 2000亿，并且市盈率 < 10 的所有公司（任务提示：使用布尔形式选取数据，阅读参考文章②③）
'''

df_company[(df_company.marketValue < 2000) & (df_company.peratio<10)]
```

```python
'''任务四：
假设其中一家公司股票的每日收盘价数据，如下'''

import numpy as np

dr = pd.date_range(start='2019-01-02', periods=100)

data = np.random.randn(100).cumsum()

close = data - np.min(data)

df = pd.DataFrame({"close": close}, index=dr)

df
```

```python
# 请计算收盘价的五日平均，并新增一列存储（任务提示：使用窗函数rolling计算移动平均，阅读参考文章④）
#  

# df["ma5day"] = df["close"].rolling(window=5).mean()
df["ma5day"] = df["close"].rolling(5).mean()
df
```
