```python

'''
导入 tushare 数据包，提取 A 股基本数据信息，取毛利率 gpr 和净利率 npr 的平均数，并分别以水球图和仪 表盘图表现出来(阅读参考文章 2、3);
'''
# 导入tushare包，需要pip安装 pip install tushare
import tushare as ts
# 获取股票基本数据
df = ts.get_stock_basics()
df.sort_index(axis=0)
```

```python
# 导入pyecharts.charts中的Bar，作水球图
from pyecharts.charts import Liquid, Page
# 导入配置项入口
from pyecharts import options as opts

from pyecharts.globals import SymbolType

from pyecharts.commons.utils import JsCode

jsfn = """function (param) {
                        return (Math.floor(param.value * 10000) / 100) + '%';
                    }"""


# A股所有股票取中位数计算毛利率
gpr = df["gpr"].mean()/100
print(gpr)
# A股所有股票取中位数计算净利率
npr = df["npr"].mean()/100
print(npr)

# 构造水球图对象
l = Liquid()
# l.add("lq", [数据1-->显示百分比的数据, 数据2--->第二个数据])
# 水球外形，有' circle', 'rect', 'roundRect', 'triangle', 'diamond', 'pin', 'arrow' 可选。
    # 默认 'circle'。
l.add("lq", [gpr, npr],shape='roundRect',is_outline_show=False ,is_animation=True, label_opts=opts.LabelOpts(
                font_size=50,
                formatter=JsCode(jsfn),
                position="inside",))
# 设置标题
l.set_global_opts(title_opts=opts.TitleOpts(title="毛利率-净利率"))

# 渲染图片
l.render_notebook()
```

```python
import numpy as np
from pyecharts.charts import Page,Gauge
from pyecharts import options as opts

gpr_int = np.round(gpr * 100)

g = Gauge()
g.add("毛利率",[("毛利率", gpr_int)], split_number=5, 
      axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(
                    color=[(0.2, "#67e0e3"), (0.8, "#37a2da"), (1, "#fd666d")], width=30
                ))
     )
g.set_global_opts(title_opts = opts.TitleOpts(title = "毛利率-仪表盘图"))
g.render_notebook()
```

```python
'''
统计 A 股各行业公司数量，并跟进行业分布数据作词云图(阅读参考文章 4、5);
'''

df_industry = df.groupby("industry")["industry"].agg("count")
df_industry = df_industry.to_frame()  
df_industry.columns = ['公司数量']
df_industry.head()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Page, WordCloud
from pyecharts.globals import SymbolType

# 提取各个行业以及其所对应的公司数量，即 行业名字：公司数量
words = list(zip(df_industry.index.values.tolist(), df_industry["公司数量"].values.tolist()))
# print(words)

# 构造词云图对象
wc = WordCloud()
# wc.add("", 数据, word_size_range=[词云图文字的大小范围])
wc.add("行业类型", words, word_size_range=[20, 100])
# 设置标题
wc.set_global_opts(title_opts=opts.TitleOpts(title="WordCloud-基本示例"))
# 渲染图片
wc.render_notebook()
```

```python
'''
利用 tushare 提取茅台数据中的最高最低价，作双折线图，茅台股票代码 600519
(阅读参考文章 1);
'''
import tushare as ts
df = ts.get_hist_data('600519',start='2018-01-01',end='2019-12-31')
df.sort_index(axis=0,inplace = True)
df.to_csv("中国茅台2018-2019年数据.csv")
df.head()
df.loc[:,"open":"low"].head()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Line
import numpy as np
import pandas as pd

# 将时间提取出来并转成list
Time = df.index.values.tolist()
# 将最高价格提取出来并转成list
high_price = df.high.values.tolist()
# 将每日最低价格提取出来并转成list
low_price = df.low.values.tolist()

# 构造折线图对象
line = Line()
# 添加横坐标数值
line.add_xaxis(Time)
# 添加纵坐标数值
line.add_yaxis("high",high_price, is_smooth=True,  
               areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
               markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")]),)
line.add_yaxis("low",low_price, is_step=True,
               markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="min")]),)
# 设置标题为折线图
line.set_global_opts(title_opts = opts.TitleOpts(title = "折线图"))
# 设置纵坐标最小值为400
line.set_global_opts(yaxis_opts=opts.AxisOpts(type_="value",min_=400,))

# 渲染图片
line.render_notebook()
```
