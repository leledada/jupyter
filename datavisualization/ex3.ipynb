```python
import pandas as pd

import matplotlib

import matplotlib.pyplot as plt

#识别图表中的中文字符
plt.rcParams['font.sans-serif'] =['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取CSV文件“task_3.1_zhihu_timeline_answer.csv”
'''
1.读取 CSV 格式文件“task_3.1_zhihu_timeline_answer.csv”，将数据命名为 zhihu_data_answer，并查看前部数据(阅读参考文章 1、2);
'''
f = open("./task_3.1_zhihu_timeline_answer.csv",encoding="utf-8") 
zhihu_data_answer = pd.read_csv(f)

# 显示知乎数据
zhihu_data_answer.head()
```

```python
'''
2.提取“task_3.1_zhihu_timeline_answer.csv”中【回答点赞数】和【回答感谢数】列数据，制成散点图， 观察两者的关系(阅读参考文章 3、4);
'''
# 提取【回答点赞数】一列数据，并转化成list
x = zhihu_data_answer["回答点赞数"].tolist()
# 提取【回答感谢数】一列数据，并转化成list
y = zhihu_data_answer["回答感谢数"].tolist()
# 调用plt，作散点图
# plt.scatter（横坐标数值，纵坐标数值，颜色，标记符号）
plt.scatter(x,y,c='r',marker='o')

# 设置图片标题
plt.title("'回答点赞数-回答感谢数'散点图")
# 设置图片横轴标签
plt.xlabel("回答点赞数")
# 设置图片纵轴标签
plt.ylabel("回答感谢数")
# 设置图片大小
plt.rcParams['figure.figsize'] = (25,15)

# 保存图片为
plt.savefig("ex3_scatter_1.png")

# 展示图片
plt.show()
```

```python
'''
提取“task_3.1_zhihu_timeline_answer.csv”中【问题回答数】和【问题关注数】列数据，制成散点图， 观察两者的关系;
'''
# 提取【问题回答数】一列数据，并转化成list
x1 = zhihu_data_answer["问题回答数"].tolist()

# 提取【问题关注数】一列数据，并转化成list
y1 = zhihu_data_answer["问题关注数"].tolist()
# 调用plt，作散点图
# plt.scatter（横坐标数值，纵坐标数值，颜色，标记符号）
plt.scatter(x1,y1,c='purple',marker='o')

# 设置图片标题
plt.title("'问题回答数-问题关注数'散点图")
# 设置图片横轴标签
plt.xlabel("问题回答数")
# 设置图片纵轴标签
plt.ylabel("问题关注数")
# 设置图片大小
plt.rcParams['figure.figsize'] = (25,15)

# 保存图片为
plt.savefig("ex3_scatter_2.png")

# 展示图片
plt.show()
```

```python
'''
扩展资料
plt 各种图形 https://www.cnblogs.com/circleyuan/p/10350167.html
matplotlib 颜色库: https://www.waitig.com/python-matplotlib数据可视化实例注解系列-设置标注字体样.html
超详解的 Matplotlib 库(数据可视化) http://www.manongjc.com/article/43132.html
'''

'''
scater() —— 绘制气泡图 

plt.scatter(x,y)
参数说明：

x:x轴上的数值
y:y轴上的数值
s:散点标记的大小
c：散点标记的颜色
cmap：将浮点数映射成颜色的颜色映射表
'''
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
a = np.random.randn(100)
b = np.random.randn(100)

plt.scatter(a,b,s = np.power(10*a+20*b,2),c=np.random.rand(100),cmap=matplotlib.cm.RdYlBu,marker="o")

plt.show()
```
