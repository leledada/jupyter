```python
'''

1.matplotlib plt.scatter(x, y, marker=None) 
xy marker 
x>o marker.

2.seaborn  sns.jointplot(x, y, data=None, kind=scatter) 
 xy  data 
data  DataFrame 
kind  scatter
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
# Matplotlib
plt.scatter(x, y,marker='o')
plt.show()
# Seaborn
df = pd.DataFrame({'x': x, 'y': y})
sns.jointplot(x="x", y="y", data=df, kind='scatter');
plt.show()
```

```python
'''



 Matplotlib  plt.plot() 
 x  x 

 Seaborn  sns.lineplot (x, y, data=None) 
 xy  data data  DataFrame 

'module' object has no attribute 'lineplot'. 
https://stackoverflow.com/questions/51846948/seaborn-lineplot-module-object-has-no-attribute-lineplot

pip install seaborn==0.9.0
'''


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
# Matplotlib
plt.plot(x, y)
plt.show()
print(sns.__version__)
# Seaborn
df = pd.DataFrame({'x': x, 'y': y})
sns.lineplot(x="x", y="y", data=df)
plt.show()
```

```python
'''


 Matplotlib  plt.hist(x, bins=10) 
 x bins  10

 Seaborn  sns.distplot(x, bins=10, kde=True) 
 x bins kde  True
 kde  False


'''


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 
a = np.random.randn(100)
s = pd.Series(a) 
# Matplotlib
plt.hist(s)
plt.show()
# Seaborn
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
```

```python
'''


 Matplotlib  plt.bar(x, height) 
 x  x height  y 

 Seaborn  sns.barplot(x=None, y=None, data=None) 
 data  DataFrame xy  data 

'''


import matplotlib.pyplot as plt
import seaborn as sns
# 
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]
# Matplotlib
plt.bar(x, y)
plt.show()
# Seaborn
sns.barplot(x, y)
plt.show()

# Seaborn,
sns.barplot(y, x)
plt.show()
```

```python
'''

 1977  (max) (min) (median)  (Q3, Q1)


 Matplotlib  plt.boxplot(x, labels=None) 
 x labels 

 Seaborn  sns.boxplot(x=None, y=None, data=None) 
 data  DataFrame xy  data 

'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 
# 0-110*4
data=np.random.normal(size=(10,4))
print(data)

lables = ['A','B','C','D']
# Matplotlib
plt.boxplot(data,labels=lables)
plt.show()
# Seaborn
df = pd.DataFrame(data, columns=lables)
sns.boxplot(data=df)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties  
# 
labels=np.array([u"","KDA",u"",u"",u"",u""])
stats=[83, 61, 95, 67, 76, 88]
# 

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
print(angles)
stats=np.concatenate((stats,[stats[0]]))
print(stats)
angles=np.concatenate((angles,[angles[0]]))
print(angles)
# Matplotlib
fig = plt.figure()
'''
describing the position of the subplot. If the three
integers are *nrows*, *ncols*, and *index* in order, the
subplot will take the *index* position on a grid with *nrows*
rows and *ncols* columns. *index* starts at 1 in the upper left
corner and increases to the right.
'''
ax = fig.add_subplot(111, polar=True)   
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
# 
font = FontProperties(fname=r"C:\Windows\Fonts\simhei.ttf", size=14)  
ax.set_thetagrids(angles * 180/np.pi, labels, FontProperties=font)
plt.show()
```
