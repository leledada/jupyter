```python
import pyecharts
import pandas as pd
import numpy as np

df = pd.read_csv("./Task_4_lianjia_sale.csv")
df
```

```python
'''
根据任务 4 所统计的上海各个区的房均价数据，利用 Pyecharts 中的 Map，以上海地图为背景作可视化展示图
(阅读参考文章 2);
'''
# df_avg = df.groupby("区域")["价格"].agg({"区域平均价格": "mean"}).sort_values(by="区域平均价格")
# df_avg.head(5)

df_avg = df.groupby("区域")["价格"].agg("mean")
df_avg = df_avg.to_frame()  
df_avg.columns = ['区域平均价格']
df_avg = df_avg.sort_values(by="区域平均价格")
df_avg
```

```python
# 导入pyecharts.charts中的Bar，作条形图
from pyecharts.charts import Map
# 导入配置项入口
from pyecharts import options as opts

# 定义区域名字
# districts = ['金山区', '奉贤区', '嘉定区', '松江区', '青浦区', '宝山区', '闵行区', '浦东新区', '普陀区', '杨浦区', '虹口区', '长宁区', '静安区', '徐汇区', '黄浦区']

# 不用自己写一遍，用index处理一下
districts = df_avg["区域平均价格"].index.tolist()
districts=[x.strip()+'新区' if x.strip() == '浦东' else x.strip()+'区' for x in districts ]
avg_price = np.round(df_avg["区域平均价格"].values).tolist()

# 地图类型，具体参考 pyecharts.datasets.map_filenames.json 文件， 在Anaconda3\Lib\site-packages\pyecharts\datasets\map_filename.json加
m = Map()
m.add("房价", [list(z) for z in zip(districts, avg_price)], "上海")
m.set_global_opts(
# 设置标题
    title_opts=opts.TitleOpts(title="上海各区房价"),
# 设置数值范围0-10万，is_piecewise标签值连续
    visualmap_opts=opts.VisualMapOpts(max_=100000, is_piecewise=False),
)

# 直接在jupytr notebook中渲染
m.render_notebook()
```

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Map


def map_base() -> Map:
    c = (
        Map()
        .add("示例指数", [list(z) for z in zip(Faker.provinces, Faker.values())], "china")
        .set_global_opts(title_opts=opts.TitleOpts(title="Map-基本示例"))
    )
    return c

map_base().render_notebook()
```

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Map


def map_guangdong() -> Map:
    c = (
        Map()
        .add("测试疫情指数", [list(z) for z in zip(Faker.guangdong_city, Faker.values())], "广东")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-广东地图"),
            visualmap_opts=opts.VisualMapOpts(),
        )
    )
    return c

map_guangdong().render_notebook()
```

```python
# 导入tushare包，需要pip安装 pip install tushare
import tushare as ts
# 获取股票基本数据
df = ts.get_stock_basics()
df.sort_index(axis=0)
```

```python
'''
导入 tushare 数据包，提取 A 股基本数据信息，利用 groupby 函数统计每个省份的公司数量，并以中国地图 为背景作相应图表(阅读参考文章 2、3、4)
'''

df_num = df.groupby("area")["name"].agg("count")
df_num = df_num.to_frame()  
df_num.columns = ['公司数量']
df_num.head()
```

```python
# 导入pyecharts.charts
from pyecharts.charts import Map
from pyecharts import options as opts

# 将df_num的索引即【area】提取出来，并转换成list
df_province = df_num.index.values.tolist()
# 将df_num的【公司数量】一列的数据值提取出来，并转换成list
df_company_num = df_num["公司数量"].values.tolist()


m = Map()
m.add("上市公司数量", [list(z) for z in zip(df_province, df_company_num)], "china")
m.set_global_opts(
    title_opts=opts.TitleOpts(title="公司地区分布图"),
    #  is_piecewise :颜色是否分段显示(False为连续渐变,True为分段)
    visualmap_opts=opts.VisualMapOpts(max_=500, is_piecewise=True),
)

m.render_notebook()
```
