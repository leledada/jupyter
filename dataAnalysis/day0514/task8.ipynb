##### 1.用pandas读取csv格式文件，并将所有字段的重复数据去除

```python
'''
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html

'''
import pandas as pd
df = pd.read_csv('.\houses.csv', index_col=0, encoding='gbk')
df = df.drop_duplicates(["title", "price",
                    "xiaoqu_name", "size",
                    "buildyear", "huxing",
                    "chaoxiang", "zhuangxiu",
                    "cenggao", "district_name",
                    "sub_district_name"])
df.head(1)
```

###### 2.请统计出总房源数量、总小区数量、总版块数量

```python
print('房源数量:{} 套'.format(len(df)))
print('总小区数量:{} '.format(len(df.xiaoqu_name.unique())))
print('总板块数量:{} '.format(len(df.sub_district_name.unique())))
print('板块:{} '.format(df.sub_district_name.unique()))
```

###### 3.请统计出有哪些不同的户型、有哪些不同的装修情况、有哪些不同的建造年份

```python
print("上海楼市户型主要有：{}".format(df.huxing.unique()))
print("上海楼市装修类型主要有: {}".format(df.zhuangxiu.unique()))
print("上海房屋建造年份主要有：{}".format(df.buildyear.unique()))
```

###### 4.请计算出上海二手房每平方米均价

```python
df['up_price'].mean()
```

###### 5.请统计出不同户型对应的房源数量、平均面积；并从大到小进行排序，判断出主流户型是什么？平均面积多大？

```python
'''
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html
'''

# df["size"].groupby(df["huxing"]).count().sort_values(ascending=[False])  # 按户型分类，数量
# df["size"].groupby(df["huxing"]).mean().sort_values(ascending=[True])  #  按户型分类，面积平均值
df_ret1= df.groupby("huxing")["size"].agg([("number","count"),("mean_size","mean")]).sort_values(by = ['number'],ascending=[False])
print(df_ret1)
print('--------------------------------')
print('主流户型/面积是：')
print(df_ret1.head(1))
```

```python
# 查看某些满足条件的行，去掉数据中的空格
df[df['huxing'].str.strip() =="9室9厅"]
```

###### 6.获取每个小区的房源数量，并按从大到小的顺序排序

```python
# df.groupby("xiaoqu_name")["title"].agg([('house_num','count')]).sort_values(by = "house_num",ascending=False)
df.groupby('xiaoqu_name')['title'].agg(['count']).sort_values(by = 'count',ascending=False)
```

###### 7.请统计出上海每个小区每平方米的均价

```python
df.groupby('xiaoqu_name')['up_price'].agg(['mean']).sort_values(by = 'mean',ascending=False)
```

###### 8.请统计出上海二手房总价排在前5位的房源，看看超级豪宅有多壕？

```python
df.sort_values(by='price', ascending=False).head(5)
```
