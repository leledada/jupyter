```python
'''
Windows环境，在 anaconda prompt中安装pyecharts， 命令（==两端不要有空格）为：pip install pyecharts==1.5.1
'''

import pyecharts
import pandas as pd
import numpy as np
# print(pyecharts.__version__)
# 读取CSV格式文件，注意文件分隔要么用反斜杠/  要么用双斜\\
'''
读取 Task_4_lianjia_sale.csv 数据(阅读参考文章 2、3)，
利用 groupby 函数，以【区域】字段进行 groupby， 同时结合 agg 聚合函数统计各个区域平均房价;
将统计所得平均房价命名为【区域平均价格】列 (阅读参考文章 4、5)
并将数据从小到大按顺序排列(提示:df.sort_values(by = “ ”) (阅读参考文章 6));
最后展示所得前 5 行数据 (将所得的 DataFrame 命名为 df_avg);
'''
# f = open("./Task_4_lianjia_sale.csv",encoding='UTF-8')
df = pd.read_csv("./Task_4_lianjia_sale.csv")
df
```

```python
'''
利用 Pyecharts 将第(1)问所得数据作条形图(阅读参考文章 7)，图表标题为“上海各区域平均价格”(提示: 最后用 render_notebook()渲染显示出图片);
'''
# 利用groupby函数，以【区域】字段进行groupby，同时结合agg聚合函数统计各个区域平均房价，将统计所得平均房价命名为【区域平均价格】
# df.groupby("被统计的列")["选择一列做运算"].agg({"新列名": "mean"}).sort_values(by="新列名")
# df_avg = df.groupby("区域")["价格"].agg({"区域平均价格": "mean"}).sort_values(by="区域平均价格")
# df_avg.head(5)

df_avg = df.groupby("区域")["价格"].agg("mean")
df_avg = df_avg.to_frame()  
df_avg.columns = ['区域平均价格']
df_avg.head()

```

```python
# 导入pyecharts.charts中的Bar，作条形图
from pyecharts.charts import Bar
# 导入配置项入口
from pyecharts import options as opts

# 将df_avg的索引即【区域】提取出来 并转换成list,作为横坐标数据
district = df_avg.index.values.tolist()
# 将【区域平均价格】数据值转成list 并将数值转成整数，作为纵坐标数据
b = df_avg["区域平均价格"].values.tolist()
avg_price = np.round(b).tolist()

# 构造柱形图对象
bar = Bar()
# 添加x轴数据
bar.add_xaxis(district)
# 添加y轴数据
bar.add_yaxis("区域平均价格", avg_price)
# 对全局进行配置,对其中的标题进行设置
bar.set_global_opts(title_opts=opts.TitleOpts(title="上海各区域平均价格"))

# 直接在jupytr notebook中渲染
bar.render_notebook()
```

```python
'''
利用 groupby 函数，以【装修类型】字段进行 groupby，同时结合 agg 聚合函数统计各个区域房源数量;

将统计所得数据命名为【房源数量】列;最后展示所得数据(将所得的 DataFrame 命名为 df_zhuangxiu);
'''
# 利用groupby函数，以【装修类型】字段进行groupby，同时结合agg聚合函数统计各个装修类型的房源数量，将统计所得数据命名为【房源数量】
# df.groupby("被统计的列")["选择一列做运算"].agg({"新列名": "count"}).sort_values(by="新列名")
# df_zhuangxiu = df.groupby("装修类型")["价格"].agg({"房源数量": "count"})
# df_zhuangxiu

df_zhuangxiu = df.groupby("装修类型")["价格"].agg("count")
df_zhuangxiu = df_zhuangxiu.to_frame()  
df_zhuangxiu.columns = ['房源数量']
df_zhuangxiu.head()
```

```python
# 导入pyecharts.charts中的Pie，作饼图
from pyecharts.charts import Pie
# 导入配置项入口
from pyecharts import options as opts

# 将df_zhuangxiu的索引即【装修类型】提取出来，并转换成list,作为横坐标数据
zhuangxiu = df_zhuangxiu.index.values.tolist()
# 将【房源数量】数据值转成list，作为纵坐标数据
house_num = df_zhuangxiu["房源数量"].values.tolist()

# 构造饼图对象
pie = Pie()
# [list(z) for z in zip(zhuangxiu, house_num)]   输出数据为[[' 毛坯 ', 5720], [' 其他 ', 5832], [' 简装 ', 24233], [' 精装 ', 34317]]
# pie.add("", [list(z) for z in zip(zhuangxiu, house_num)])
pie.add('装修占比', [list(z) for z in zip(zhuangxiu, house_num)],
        radius=["10%", "75%"], #设置内径外径，第一项内径，第二项内径
        rosetype="area", #玫瑰图有两种类型,radius area
        is_clockwise=False,   # 饼图的扇区是否是顺时针排布。默认是True
        label_opts=opts.LabelOpts(is_show=True))
# 设置标题
pie.set_global_opts(title_opts=opts.TitleOpts(title="装修类型占比情况-玫瑰图"))

# 设置系列格式formatter="{b}: {d}"  d代表百分比，c代表数值
#  饼图、雷达图 : a（系列名称），b（数据项名称），c（数值）, d（百分比）
pie.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}"))

# 直接在jupytr notebook中渲染
pie.render_notebook()
```

```python
# https://pyecharts.org/#/zh-cn/basic_charts?id=pie%ef%bc%9a%e9%a5%bc%e5%9b%be

from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Pie


def pie_base() -> Pie:
    c = (
        Pie()
        .add("", [list(z) for z in zip(Faker.choose(), Faker.values())])
        .set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

pie_base().render_notebook()
```

```python
def pie_set_colors() -> Pie:
    c = (
        Pie()
        .add("", [list(z) for z in zip(Faker.choose(), Faker.values())])
        .set_colors(["blue", "green", "yellow", "red", "pink", "orange", "purple"])
        .set_global_opts(title_opts=opts.TitleOpts(title="Pie-设置颜色"))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

pie_set_colors().render_notebook()
```

```python
# 引包，执行JavaScript代码
from pyecharts.commons.utils import JsCode

def pie_multiple_base() -> Pie:
    fn = """
    function(params) {
        if(params.name == '其他')
            return '\\n\\n\\n' + params.name + ' : ' + params.value + '%';
        return params.name + ' : ' + params.value + '%';
    }
    """

#通过定义JavaScript回调函数自定义标签
    def new_label_opts():
        return opts.LabelOpts(formatter=JsCode(fn), position="center")

    c = (
        Pie()
        .add(
            "",
            [list(z) for z in zip(["剧情", "其他"], [25, 75])],
            center=["20%", "30%"],
            radius=[60, 80],
            label_opts=new_label_opts(),
        )
        .add(
            "",
            [list(z) for z in zip(["奇幻", "其他"], [24, 76])],
            center=["55%", "30%"],
            radius=[60, 80],
            label_opts=new_label_opts(),
        )
        .add(
            "",
            [list(z) for z in zip(["爱情", "其他"], [14, 86])],
            center=["20%", "70%"],
            radius=[60, 80],
            label_opts=new_label_opts(),
        )
        .add(
            "",
            [list(z) for z in zip(["惊悚", "其他"], [11, 89])],
            center=["55%", "70%"],
            radius=[60, 80],
            label_opts=new_label_opts(),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Pie-多饼图基本示例"),
            legend_opts=opts.LegendOpts(
                type_="scroll", pos_top="20%", pos_left="80%", orient="vertical"
            ),
        )
    )
    return c

pie_multiple_base().render_notebook()
```

```python
def pie_scroll_legend() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [
                list(z)
                for z in zip(
                    Faker.choose() + Faker.choose() + Faker.choose(),
                    Faker.values() + Faker.values() + Faker.values(),
                )
            ],
            center=["40%", "50%"],
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Pie-Legend 滚动"),
            legend_opts=opts.LegendOpts(
                type_="scroll", pos_left="80%", orient="vertical"
            ),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

pie_scroll_legend().render_notebook()
```
