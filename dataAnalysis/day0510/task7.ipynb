```python
'''
任务一：
利用 pandas 读取 csv 格式数据文件，去除其中的重复值，并展示前部数据（阅读参考文章①②）
'''
import pandas as pd
df = pd.read_csv('.\jobs.csv')
df.drop_duplicates(inplace=True)
df.head()
```

```python
'''
任务二
统计数据表中的招聘岗位都来自于哪些网站（阅读参考文章③④）

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html

pandas.Series.value_counts

Series.value_counts(self, normalize=False, sort=True, ascending=False, bins=None, dropna=True)
参数：
normalize  bool, default False
If True then the object returned will contain the relative frequencies of the unique values. 如果为True，则以百分比的形式显示

sort bool, default True
Sort by frequencies. 排序,默认是排序了的

ascending bool, default False
Sort in ascending order. 默认是降序

bins int, optional
Rather than count values, group them into half-open bins, a convenience for pd.cut, only works with numeric data.
只适用数字数据

dropna bool, default True
Don’t include counts of NaN. 默认删除空值

Returns
Series  返回值也是 Series
 
'''

# df['source']
df['source'].value_counts()
```

```python
'''
任务三：
统计数据分析师这个岗位，对于工作经验的要求
'''
df['experience'].value_counts()
```

```python
'''
任务四：
统计该地区数据分析岗位对于教育背景的要求
'''
df['education'].value_counts()
```

```python
'''
任务五：
请统计出不同类型公司的职位数量（阅读参考文章③④）
'''

df['company_type'].value_counts()
```

```python
'''
任务六：
所有职位的平均工资是多少
'''

df['salary2'].mean()
```

```python
'''
任务七：
请统计平均薪资（即 salary2）排在前 10 位的招聘公司及金额（阅读参考文章⑤⑥）
'''
# 写法一
df.groupby(['company'])['salary2'].mean().sort_values(ascending=False).head(10)

```

```python
#  写法2， 比写法1好理解一点？
df['salary2'].groupby(df['company']).mean().sort_values(ascending=False).head(10)
```

```python
# 写法3, 这个数据有一点不一样呢~~~
df.groupby('company').mean().sort_values(by=['salary2'],ascending=False).head(10)
```

```python
'''
任务八：
请统计发布职位数排在前 10 位的招聘公司及其职位数量（阅读参考文章⑤⑥）
'''

df.groupby("company")["title"].count().sort_values(ascending=False).head(10)
```

```python
df['title'].groupby(df['company']).count().sort_values(ascending=False).head(10)
```

```python

df.groupby("company").count().sort_values(by=['title'],ascending=False).head(10)
```

```python
df[df['company'] == "百度在线网络技术(北京)有限公司"]
```

```python
'''
任务九：
请统计出平均工资排在前 10 位的公司类型及平均工资金额（阅读参考文章⑤⑥）
'''
#  直接用去掉麻烦列的方法
df.groupby('company_type').mean().sort_values(by=['salary2'],ascending=False).head(10)
```

```python
#  方法二
df.groupby('company_type')['salary2'].mean().sort_values(ascending=False).head(10)
```
